## Loops


```python
# Print "Hello world" 10 times
# While loop
counter = 0 
while counter < 10: # while condition:
    print("Hello, world!") # outcome
    counter +=1                           # termination condition
    
```

```python
# Create a program that prints a countdown from 5 and then says "blast off"
counter = 5
while counter <= 5 and counter > 0:
    print(counter)
    counter -= 1
    if counter == 0:
        print("blast off!")
```
